```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
import streamlit as st

df = pd.read_csv("~data-proj-5\\data-proj-5\\starter\\supermarket.csv")
df.head()
```

```python
#
topTenStores = df.sort_values(by=['store_sales'], ascending = False)
topTenStores.head(10)
```

```python
topTenAreas = df.groupby('store_area')['store_sales'].sum().sort_values(ascending = False).head(10)
tenAreas = topTenAreas.to_frame().reset_index()
sns.barplot(data = tenAreas, x = 'store_area', y = 'store_sales')
```

```python
avgDailyCust = df['daily_customer_count'].mean()
print("The average daily customer amount is " + str(avgDailyCust))
```

```python
totalSales = df['store_sales'].sum()
print("The total amount of sales is " + str(totalSales))
```
